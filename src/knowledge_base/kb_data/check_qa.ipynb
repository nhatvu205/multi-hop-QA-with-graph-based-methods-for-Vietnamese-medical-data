```python
import json
```

```python
with open('knowledge_base.json', "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
# ...existing code...
qa_id = ['qa_2594']
def find_matches(obj, key, values):
    if isinstance(values, str):
        values = [values]
    matches = []
    if isinstance(obj, dict):
        v = obj.get(key)
        if v in values:
            matches.append(obj)
        for child in obj.values():
            matches.extend(find_matches(child, key, values))
    elif isinstance(obj, list):
        for item in obj:
            matches.extend(find_matches(item, key, values))
    return matches

matches = find_matches(data, "qa_id", qa_id)
if matches:
    for i, m in enumerate(matches, 1):
        print(f"--- Kết quả {i} ---")
        print(json.dumps(m, ensure_ascii=False, indent=2))
else:
    print(f"Không tìm thấy qa_id = {qa_id}")
# ...existing code...
```
